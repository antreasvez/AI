### Αλγόριθμος παραγωγής λαβυρίνθων
```
1. Αρχικά όλα τα κελιά περιέχουν εμπόδια και η λίστα των τοίχων είναι κενή.
2. Ένα κελί επιλέγεται ως μέρος του λαβυρίνθου. Οι τοίχοι του κελιού προστίθενται στη λίστα των τοίχων.
3. Όσο η λίστα των τοίχων δεν είναι κενή:
      1. Επιλέγεται τυχαία ένας τοίχος από τη λίστα. Εαν μόνο ένα από τα (δύο) κελιά που χωρίζει ο τοίχος είναι μέρος του λαβυρίνθου, τότε:
            1. Ο τοίχος και το κελί που δεν είναι μέρος του λαβυρίνθου, γίνονται μέρος του λαβυρίνθου.
            2. Οι τοίχοι των νέων κελιών που προστέθηκαν στο λαβύρινθο προστίθενται στη λίστα των τοίχων.
      2. Ο τοίχος αφαιρείται από τη λίστα των τοίχων
```
*ως τοίχοι ενός κελιού θεωρούνται όσα απο τα 4 γειτονικά κελιά (όχι τα διαγώνια) περιέχουν εμπόδιο

````
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from IPython.display import HTML

class visualization:
    def __init__(self, S, F):
        '''
          Η μέθοδος αυτή αρχικοποιεί ένα αντικείμενο τύπου visualization.
          Είσοδος: 
          -> S: το σημείο εκκίνσης της αναζήτησης
          -> F: το σημείο τερματισμού
        '''
        self.S = S
        self.F = F
        self.images = []
    
    def draw_step(self, grid, frontier, expanded_nodes):
        '''
          Η συνάρτηση αυτή καλείται για να σχεδιαστεί ένα frame στο animation (πρακτικά έπειτα από την επέκταση κάθε κόμβου)
          Είσοδος: 
          -> grid: Ένα χάρτης τύπου grid
          -> frontier: Μια λίστα με τους κόμβους που ανήκουν στο μέτωπο της αναζήτησης
          -> expanded_nodes: Μια λίστα με τους κόμβους που έχουν ήδη επεκταθεί
          Επιστρέφει: None
          Η συνάρτηση αυτή πρέπει να καλεστεί τουλάχιστον μια φορά για να μπορέσει να σχεδιαστει ένα animation (πρεπεί το animation να έχει τουλάχιστον ένα frame).
        '''
        image = np.zeros((grid.N, grid.N, 3), dtype=int)
        image[grid.grid == 0] = [255, 255, 255]
        image[grid.grid == 1] = [0, 0, 0]
        
        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        self.images.append(image)
    
    def add_path(self, path):
        '''
          Η συνάρτηση αυτή προσθέτει στο τελευταίο frame το βέλτιστο μονοπάτι.
          Είσοδος:
          -> path: Μια λίστα η όποια περιέχει το βέλτιστο μονοπάτι (η οποία πρέπει να περιέχει και τον κόμβο αρχή και τον κόμβο στόχο)
          Έξοδος: None
        '''
        for n in path[1:-1]:
            image = np.copy(self.images[-1])
            image[n] = [66, 221, 245]
            self.images.append(image)
        for _ in range (100):
            self.images.append(image)
        
    def create_gif(self, fps = 30, repeat_delay = 2000):
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        fig = plt.figure()
        plt.axis('off')
        ims = []
        for img in self.images:
            img = plt.imshow(img)
            ims.append([img])
        ani = animation.ArtistAnimation(fig, ims, interval=1000//fps, blit=True, repeat_delay= repeat_delay)
        plt.close(fig)
        return ani
        
    def save_gif(self, filename, fps = 30):
        '''
            Η συνάρτηση αυτή ξαναδημιουργεί και αποθηκεύει το animation σε ένα αρχείο.
            Είσοδος:
            -> Το όνομα του αρχείου με κατάληξη .gif
            Έξοδος: (None)
        '''
        ani = self.create_gif(fps)
        writer = PillowWriter(fps= fps)
        ani.save(filename, writer=writer)

    def show_gif(self, fps= 30, repeat_delay = 2000):
        '''
            Η συνάρτηση αυτή εμφανίζει inline το animation.
            Είσοδος:
            -> fps: τα frames per second
            Έξοδος: Το αντικείμενο που παίζει το animation
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        ani = self.create_gif(fps, repeat_delay)
        # return HTML(ani.to_html5_video())
        return HTML(ani.to_jshtml())

    def show_last_frame(self):
        '''
            Η μέθοδος αυτή εμφανίζει inline το τελευταίο frame που έχει δημιουργήθει.
            Είσοδος:
            Έξοδος: Το αντικείμενο που εμφανίζει την εικόνα.
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        else:
            plt.imshow(self.images[-1])


class EmptyStackOfImages(Exception):
    pass
````

````
%matplotlib inline
import numpy as np
from queue import LifoQueue
from random import shuffle, uniform, choice
import matplotlib.pyplot as plt

class grid:
    def __init__(self, N, S, F, distribution = lambda : 0):
        
        ## Make sure start and end are within the grid
        assert N > 2
        assert S[0] < N
        assert S[1] < N
        assert F[0] < N
        assert F[1] < N

        assert S[0] > 0
        assert S[1] > 0
        assert F[0] > 0
        assert F[1] > 0
        
        self.distribution = distribution
        self.weight_dict = {}
  
        self.N = N

        ## Initialize grid with obstacles
        self.grid = np.ones((N, N), dtype=np.int32)
        
        ## Start and end position have no obstacles
        self.grid[S]=0
        self.weight_dict[S] = 0

        # Don't add F to the labyrinth. It will be added in the end
        #self.grid[F]=0       

        obstacle_free_points = {S, F}
        
        ### YOUR CODE HERE
        wall_list = []
        wall_list += self.neighboring_walls(S)
        
        # Create a random maze
        while wall_list:
          curr = choice(wall_list)
          wall_list.remove(curr)

          # Vertical wall
          if curr[0]+1 < self.N and curr[0]-1 >= 0 and self.grid[curr] == 1: 
            # Check whether there is only one unvisited horizontal side of this wall
            if self.grid[(curr[0]+1, curr[1])] == 1 - self.grid[(curr[0]-1, curr[1])]:
              # Mark unvisited cell as part of the maze
              for x in [(curr[0]+1, curr[1]), (curr[0]-1, curr[1])]:
                if self.grid[x] == 1:
                  wall_list += self.neighboring_walls(x)
                  self.grid[x] = 0
                  self.weight_dict[x] = self.distribution() 
              # Mark wall as part of the maze
              self.grid[curr] = 0
              self.weight_dict[curr] = self.distribution()
            
          # Horizontal wall
          if curr[1]+1 < self.N and curr[1]-1 >= 0 and self.grid[curr] == 1:
            # Check whether there is only one unvisited vertical side of this wall
            if self.grid[(curr[0], curr[1]+1)] == 1 - self.grid[(curr[0], curr[1]-1)]:
              # Mark unvisited cell as part of the maze
              for x in [(curr[0], curr[1]+1), (curr[0], curr[1]-1)]:
                if self.grid[x] == 1:
                  wall_list += self.neighboring_walls(x)
                  self.grid[x] = 0
                  self.weight_dict[x] = self.distribution()
              # Mark wall as part of the maze
              self.grid[curr] = 0  
              self.weight_dict[curr] = self.distribution()
    
        # F is either part of the labyrinth or a wall
        # In any case make it part of the labyrinth
        self.grid[F] = 0
        self.weight_dict[F] = self.distribution()

    # Checks whether position is within bounds      
    def within_bounds(self, pos):
      return pos[0] >= 0 and pos[0] < self.N and pos[1] >= 0 and pos[1] < self.N

    # Returns all neighbors within bounds
    def neighbors(self, curr):
      return [x for x in [(curr[0]-1, curr[1]), 
                          (curr[0]+1, curr[1]),
                          (curr[0], curr[1]-1),
                          (curr[0], curr[1]+1)] 
              if self.within_bounds(x)]

    # Returns neighbors that are not part of the maze
    def neighboring_walls(self, pos):
      return [x for x in self.neighbors(pos) if self.grid[x] == 1]

    # Returns neighbors that are part of the maze
    def neighboring_cells(self, pos):
      return [x for x in self.neighbors(pos) if (self.grid[x] == 0)]

    def draw_map(self, S=None, F=None, path=None):
        image = np.zeros((self.N, self.N, 3), dtype=int)
        image[self.grid == 0] = [255, 255, 255]
        image[self.grid == 1] = [0, 0, 0]
        if S:
            image[S] = [50, 168, 64]
        if F:
            image[F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()
````

````
# g(n) is the distance between two cells if there is no random distribution on the grid

# Flag use_distribution: 
#           False -> Distribution is not taken into consideration 
#           True  -> g(n) doesn't do anything and distances are calculated
class pathfinder:
    def __init__(self, S, F, grid, c, h, use_distribution = False):
        self.S = S
        self.F = F
        self.grid = grid
        self.cost = c
        self.heuristic = h
        #for visualization 
        self.vis = visualization(S, F)
        self.path=[]
        self.use_distribution = use_distribution
        self.dict_g = {}       # dictionary with distances 
    
    # Calculates the value of function f for a cell
    def calc_f(self, n):
        return self.dict_g[n] + self.heuristic(n[0], n[1])

    # Find minimum element in frontier and return its index
    def find_min_index(self, frontier):
      minimum = self.calc_f(frontier[0])
      index = 0
      for i in range(len(frontier)):
        cost_i = self.calc_f(frontier[i])
        if cost_i < minimum:
          index = i
          minimum = cost_i
      return index
        
    def find_path(self):
      self.path = []          # path from start to finish
      parent = {}             # dictionary of parents
      frontier = []           # list nodes to be visited
      expanded_nodes = []     # list of expanded nodes      
      self.dict_g[self.S] = 0
      curr = self.S
      
      while (curr != self.F):
        # Find children of current
        children = self.grid.neighboring_cells(curr)

        # Make neighbor cells that have not been added to frontier, 
        # children of curr and add them to frontier
        for i in children:
          # new_cost is the distance of i from the start if we follow the current path
          if self.use_distribution:
            new_cost = self.dict_g[curr] + self.grid.weight_dict[i]
          else:
            new_cost = self.dict_g[curr] + self.cost(i[0], i[1])
          # If we haven't reached i yet, add it to the lists/dictionaries
          if i not in self.dict_g:
            frontier.append(i)
            self.dict_g[i] = new_cost
            parent[i] = curr
          # Change distance if the new path is better
          elif new_cost < self.dict_g[i]:
            self.dict_g[i] = new_cost
            parent[i] = curr

        # Mark current cell as expanded
        expanded_nodes.append(curr)

        # Add frame to animation
        self.vis.draw_step(self.grid, frontier, expanded_nodes)

        # Index is the position of the cell with minimum f
        index = self.find_min_index(frontier)

        # Pop the closest cell to destination
        curr = frontier.pop(index)
      
      # Traverse parent to find the path   
      while (curr != self.S):
        self.path = [curr] + self.path
        curr = parent[curr]
        
      # Add S to the path    
      self.path = [self.S] + self.path

      # Add path to animation
      self.vis.add_path(self.path)
      return (np.size(self.path),np.size(expanded_nodes))
        
    def get_path(self):
      return self.path
````

#### Παραδείγματα εκτέλεσης με διαφορετικές ευριστικές

Για τις ευριστικές που θα χρησιμοποιηθούν παρακάτω θα ισχύουν τα εξής


*   Για $g(n) = 1, h(n) = 0$ έχουμε τον αλγόριθμο DFS. Λόγω της σταθερής τιμής του κόστους μετάβασης σε οποιονδήποτε κόμβο, είναι στην συγκεκριμένη περίπτωση ισοδύναμος με τον αλγόριθμο Dijkstra, ο οποίος πάντα βρίσκει τη βέλτιστη διαδρομή. 

*   Για $g(n) = 1, h(n) = {manhattan, euclidean}$ προκύπτει ο A*. Η ευριστική manhattan είναι πάντα ίση με τα ελάχιστα, θεωρητικά, βήματα που μπορούν να γίνουν για να φτάσει στο τέλος οπότε ικανοποιείται το κριτήριο και πάντα βρίσκει το βέλτιστο μονοπάτι. Η ευριστική euclidean 
υπολογίζει την απόσταση από το τέλος ως πραγματικό αριθμό, ο οποίος θα είναι μικρότερος από την τιμή της manhattan (ως μήκος υποτείνουσας του 
ορθογωνίου τριγώνου που σχηματίζεται αν σχεδιάσουμε την διαδρομή manhattan και την ευθεία που ενώνει το τρέχον σημείο με το τέλος).

*   Για $g(n) = 0, h(n) = {manhattan, euclidean}$ προκύπτει ο αλγόριθμος Best-First. Δεδομένου ότι ακολουθεί τον κόμβο του μετώπου με ελάχιστη ευριστική, θα βρει σίγουρα διαδρομή από το S στο F, αλλά όχι αναγκαστικά την βέλτιστη. Στη δική μας υλοποίηση βέβαια, του λαβυρίνθου, υπάρχει μοναδική διαδρομή από την αρχή στο στόχο, οπότε πάντα θα βρίσκει αυτήν.

````
from random import uniform, betavariate, triangular
import math

def manhattan(dest, x, y):
  return abs(dest[0] - x) + abs(dest[1] - y)

def euclidean(dest, x, y):
  return math.sqrt((dest[0] - x)**2 + (dest[1] - y)**2)

def more_distributions1(dest,x,y):
    return max(abs(dest[0] - x), abs(dest[1] - y))

def more_distributions2(dest,x,y):
    return min(abs(dest[0] - x), abs(dest[1] - y))

def func(g,f,map):
    x1 = []
    x2 = []
    for i in range(10):
        N = (i+1)*10
        S = (1,1)
        F = (N-2, N-2)
        pf = pathfinder(S, F, map[i], g, lambda x,y: f(F,x,y), True)
        temp = pf.find_path()
        x1 += [temp[0]]
        x2 += [temp[1]]
    return (x1,x2)
````
